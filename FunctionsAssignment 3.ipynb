**1.What is the difference between a function and a method in Python?**

In Python, both functions and methods are callable objects, but they have key differences:

Definition:



1.Function: A function is a block of reusable code that performs a specific task. It is defined using the def keyword and can be called independently, outside of any class or object.






- Function: Can be defined globally or locally, and is not bound to any object.






2.Method: A method is a function that is associated with an object (or class). It is called on an instance (or class) and typically operates on that instance's data. Methods are defined inside a class, and the first parameter of a method is always self (referring to the instance of the class).






- Method: Always associated with an object (instance method) or class (class method). It operates on data specific to that object or class.






Exmaple -

```python
# Function Example
def greet(name):
    return f"Hello, {name}"

# Method Example (within a class)
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):  # Method
        return f"Hello, {self.name}"

# Using the function
print(greet("Alice"))  # Output: Hello, Alice

# Using the method
person = Person("Bob")
print(person.greet())  # Output: Hello, Bob
```

**2. Explain the concept of function arguments and parameters in Python?**

In Python, function arguments and parameters are used to pass data to functions so that they can operate on it. Here’s a breakdown of these terms:

**3. What are the different ways to define and call a function in Python?**

In Python, there are several ways to define and call a function. Let's go through the different approaches and provide examples for each.

1.Defining a Simple Function A function in Python is defined using the def keyword, followed by the function name, parentheses, and a colon. The block of code inside the function is indented.




- Example:

```python
def greet():
    print("Hello, World!")

# Calling the function
greet()  # Output: Hello, World!
```

2.Defining a Function with Parameters You can define a function that takes parameters. These parameters allow you to pass data into the function.
- Example:

```python
def greet(name):
    print(f"Hello, {name}!")

# Calling the function with an argument
greet("Alice")  # Output: Hello, Alice!
```

3.Defining a Function with Return Value A function can also return a value using the return keyword. This allows the function to produce a result that can be used elsewhere in your program.
- Example:

```python
def add(a, b):
    return a + b

# Calling the function and storing the result
result = add(3, 5)
print(result)  # Output: 8
```

4.Defining a Function with Default Parameters You can define a function with default parameter values. If the argument is not provided when calling the function, the default value will be used.
- Example:

```python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

# Calling the function with and without the second argument
greet("Alice")  # Output: Hello, Alice!
greet("Bob", "Good evening")  # Output: Good evening, Bob!
```

5.Defining a Function with Variable-length Arguments You can define functions that accept a variable number of arguments using args (for positional arguments) and *kwargs (for keyword arguments).
- Example with *args (Variable Positional Arguments):

```python
def add(*args):
    return sum(args)

# Calling the function with different numbers of arguments
print(add(1, 2, 3))  # Output: 6
print(add(4, 5))  # Output: 9
```

Example with **kwargs (Variable Keyword Arguments):

```python
def greet(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

# Calling the function with different keyword arguments
greet(name="Alice", age=30)  # Output: name: Alice, age: 30
```

6.Lambda Functions (Anonymous Functions) Python also allows you to define small, anonymous functions using the lambda keyword. These functions can have any number of arguments but only one expression.
- Example:

```python
# Defining a lambda function
add = lambda a, b: a + b

# Calling the lambda function
print(add(3, 4))  # Output: 7
```

7.Defining a Recursive Function A recursive function is one that calls itself. It is commonly used for tasks like calculating factorials or traversing data structures (like trees).
- Example:

```python
def factorial(n):
    if n == 0:  # Base case
        return 1
    else:  # Recursive case
        return n * factorial(n - 1)

# Calling the recursive function
print(factorial(5))  # Output: 120
```

**4. What is the purpose of the `return` statement in a Python function?**

Key Points about the return Statement:



1.Terminates Function Execution: When the return statement is encountered, the function execution stops immediately, and the program control is transferred back to the point where the function was called.

**6. Explain the concept of generators in Python and how they are defined.**

In Python, generators are a type of iterable, like lists or tuples, but with a key difference: they generate items on the fly (lazily) rather than storing them in memory all at once. This makes generators memory-efficient, especially when dealing with large datasets or infinite sequences.

Key Features of Generators:
1. Lazy Evaluation:

- Generators compute values only when requested, using a special keyword yield. This allows them to handle large data efficiently since they don’t generate the entire sequence at once.


2. Memory Efficiency:

- Unlike lists, which store all their elements in memory, generators produce
 items one at a time, and only as needed. This makes them suitable for large or infinite data sequences.


3. State Retention:

-  When a generator function is called, it returns an iterator that generates
  items one by one. It remembers its state between iterations, so the execution of the function is paused when yield is encountered and resumes where it left off when the next item is requested.


  
  4.Syntax:

- Generators are defined using functions with the yield keyword, rather than
  return. A generator can yield multiple values, one at a time.


Defining a Generator:


Generators are defined like regular functions, but instead of using return to return a value, the yield keyword is used. When yield is encountered, the function pauses and sends a value back to the caller. The state of the function (including local variables) is saved, and the function can resume where it left off when next() is called again.

Example 1: Simple Generator


```python
# Generator function that yields numbers from 1 to 3
def my_generator():
    yield 1
    yield 2
    yield 3

# Creating the generator object
gen = my_generator()

# Iterating through the generator
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
# print(next(gen))  # Raises StopIteration
```

Example 2: Generator for a Range of Numbers (Using yield in place of return)

```python
# Generator to generate a range of numbers
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Using the generator
counter = count_up_to(5)

# Iterating through the generator
for number in counter:
    print(number)
```

Example 3: Infinite Generator
A generator can also handle infinite sequences since it computes values on the fly and doesn’t store them.

```python
# Infinite generator to yield odd numbers
def infinite_odds():
    num = 1
    while True:
        yield num
        num += 2

# Using the generator
odd_numbers = infinite_odds()
for _ in range(5):
    print(next(odd_numbers))  # Output: 1 3 5 7 9
```

**7. What are the advantages of using generators over regular functions?**

Using generators over regular functions in Python offers several advantages, particularly in terms of memory efficiency, performance, and handling large datasets. Here's a breakdown of the key benefits:

1. Memory Efficiency
Generators produce values on the fly, meaning they do not store the entire sequence of values in memory at once.

- This is particularly useful when working with large datasets or infinite sequences, as generators only store the current state and yield one item at a time. This significantly reduces memory usage.

2. Lazy Evaluation

- Generators use lazy evaluation, meaning they compute values only when
 requested (using next() or in a loop). This is useful when you need to work with large data but don't need all the values at once.

3. Better Performance in Certain Use Cases
- Generators can provide better performance for specific tasks, such as   looping  over a sequence of data where you only need to process one item at a
  time. They eliminate the overhead of storing and retrieving large collections in memory.

4. Infinite Sequences
- Generators can represent infinite sequences, as they generate values one at a
  time. Regular functions would not be able to handle infinite data because they would need to generate and store the entire sequence in memory, which is impossible for infinite data.

5. Cleaner and More Readable Code
- Generators help in writing more concise and readable code compared to
  traditional methods for handling large datasets.


6. State Retention Between Calls
- Generators remember their state between yield calls. This allows them to  
  pause execution and resume from the point where they left off, without the need to explicitly maintain state (e.g., via global variables or additional data structures).

7. Reduced Code Complexity in Iterative Processes
   When dealing with complex iterative processes (e.g., reading from a file line by line or processing data streams), generators can simplify the code by handling the iteration and state management implicitly.

**9. Explain the purpose and usage of the `map()` function in Python.**

The map() function in Python is a built-in function that applies a given function to all the items in an iterable (such as a list, tuple, or string) and returns an iterator (or map object) that yields the results of applying the function to each item in the iterable. The map() function is especially useful when you need to transform or process each element of an iterable in a consistent way.

Syntax of map():


- function: The function that will be applied to each item in the iterable. This can be a normal function or a lambda function.
- iterable: The iterable whose items will be passed to the function.
- You can pass multiple iterables to map(), in which case the function should accept that many arguments (one for each iterable).



Key Points:



- map() returns a map object, which is an iterator that produces the results one by one.
- The map object is lazy, meaning the transformation happens when you iterate over the map object, not when it is created. If you want to get the results as a list, you need to explicitly convert the map object using list() or tuple().


Advantages of Using map():



- Concise and Readable: Using map() can make the code more concise and readable, especially for simple transformations or calculations applied to each item in an iterable.
- Functional Programming: map() fits well with the functional programming paradigm, where functions are treated as first-class citizens and can be passed around as arguments to other functions.
- Efficient for Large Data: Since map() returns a lazy iterator (map object), it is memory-efficient, especially for processing large datasets. The transformation happens only when needed, without the need to store intermediate results.

Although they share similarities in that they apply a function to an iterable (or multiple iterables), each has a distinct purpose and behavior.

1. map() Function
- Purpose: The map() function is used to apply a function to each item of an iterable (or multiple iterables) and returns an iterator (or map object) with the results.

2. reduce() Function
- Purpose: The reduce() function is used to reduce an iterable to a single accumulated result by applying a function that takes two arguments. The function is applied cumulatively to the items of the iterable, reducing it to a single value.

- Usage: It's typically used when you need to aggregate or combine elements from an iterable into a single result (e.g., summing all numbers, finding the product of all elements).


3. filter() Function


- Purpose: The filter() function is used to filter out elements from an iterable based on a condition (i.e., the function you provide should return True or False for each element). Only the elements that satisfy the condition (i.e., the function returns True) will be included in the result.

- Usage: It's typically used when you need to filter elements from an iterable based on some condition.

**1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**

```python
def sum_even_numbers(numbers):
    # Initialize the sum to 0
    total = 0
    # Loop through each number in the list
    for num in numbers:
        # Check if the number is even
        if num % 2 == 0:
            total += num  # Add even number to total
    return total

# Example usage:
numbers = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(numbers)
print("Sum of even numbers:", result)
```

**4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

        


```python
def is_prime(n):
    # Check if the number is less than 2, since primes are greater than 1
    if n < 2:
        return False
    # Check for factors from 2 to sqrt(n)
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False  # Found a factor, so it's not prime
    return True  # No factors found, so it's prime

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is prime")
```

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms  # Number of terms in the Fibonacci sequence
        self.current = 0    # Current index in the sequence
        self.a, self.b = 0, 1  # The first two Fibonacci numbers

    def __iter__(self):
        return self  # Return the iterator object itself

    def __next__(self):
        if self.current >= self.terms:
            raise StopIteration  # Stop iteration when the required number of terms is reached
        else:
            # Return the current Fibonacci number and update a and b for next term
            fib_num = self.a
            self.a, self.b = self.b, self.a + self.b  # Update for next Fibonacci numbers
            self.current += 1
            return fib_num

# Example usage:
fibonacci = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers
for num in fibonacci:
    print(num)
```

**6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i  # Yield the power of 2 for the current exponent

# Example usage:
for power in powers_of_2(5):  # Generate powers of 2 up to 2^5
    print(power)
```

**9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit**

```python
# List of temperatures in Celsius
celsius_temps = [0, 10, 20, 30, 40, 50]

# Convert Celsius to Fahrenheit using map and a lambda function
fahrenheit_temps = list(map(lambda c: (9/5) * c + 32, celsius_temps))

# Print the list of temperatures in Fahrenheit
print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_temps)
```

**10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
# Function to check if a character is a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Given string
input_string = "Hello, World!"

# Use filter() to remove vowels
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Print the result
print("Original String:", input_string)
print("String without vowels:", filtered_string)
```
